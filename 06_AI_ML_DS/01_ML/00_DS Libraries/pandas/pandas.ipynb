# Initalization

```python
import pandas as pd
```

## Load Series

```python
# Series
a = [1,2,2]
dataset = pd.Series(a)
print(dataset)
```

```python
# Series with labels
dataset = pd.Series(a, index=["No.1", "No.2", "No.3"])
print(dataset)
```

## Load Dataframe

```python
# Dataframe
a = {
    "name":["Mr A", "Mr B", "Mr C"],
    "color":["Red", "Green", "Pink"]
}
dataset = pd.DataFrame(a)
print(dataset)
```

### iloc - Integer location

```python
# To get row 2 data
dataset.iloc[1]
```

```python
# to get range of columns
dataset.iloc[0:5]
```

```python
# To acquire specific cell data
dataset.iloc[1,4]
```

```python
# get data of all columns, excluding last coulmn
dataset.iloc[:,:-1]
```

```python
# get data of Only last coulmn
dataset.iloc[:,-1]
```

### iterrows()

```python
for index,row in dataset.iterrows():
    print(index, row["Name"])
```

### Data slicing with filter

```python
# get coulmns with speed > 90
dataset.loc[dataset["Speed"]>90]
```

```python
# Get coulmns with Type2 = Poison
dataset.loc[dataset["Type 2"]=="Poison"]
```

# Sorting

# Editing data

```python
# Add columns
dataset["Power"] = dataset["HP"] + dataset["Attack"]
dataset.head(3)
```

# Save dataset in new file

```python
dataset.to_csv("newdata.csv")
dataset.to_excel("newdata.xlsx")
dataset.to_csv("newdata.txt", index=False, sep="\t")
```

# Checking NaN values

```python
dataset.isna().any()
```

```python
# Remove entire column 
dataset[dataset["Type 2"].notna()]
```

```python
# for numeric columns take mean and replace empty value with mean or median
dataset["Speed"].fillna(dataset["Speed"].mean())
dataset["Speed"].fillna(dataset["Speed"].median())
```

# Mapping certain data to another value

```python
Generation = set(dataset['Generation'])
dataset['Generation'] = dataset['Generation'].map({1:"one", 2:"two", 3:"three", 4:"four", 5:"Five", 6:"Six"}).astype(str)
print(dataset.head(2))
```
