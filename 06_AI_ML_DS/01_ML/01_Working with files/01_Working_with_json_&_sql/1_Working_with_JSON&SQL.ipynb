```python
# This library Connects python to mySQL DB
import mysql.connector

# Specifiying the mySQL DB
conn = mysql.connector.connect(host='localhost', user='root', password='', database='world')
```

```python
# Display city table from world database
pd.read_sql_query("select * from city", conn).sample(2)
```

```python
# Applying sql queries :-
pd.read_sql_query("select * from city where CountryCode like 'IND'", conn).sample(2)
```

```python
pd.read_sql_query("select * from country where LifeExpectancy > 60", conn).sample(2)
```
