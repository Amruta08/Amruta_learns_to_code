```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.allupost.com/blog/best-companies-in-india-to-work/"

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36',
    'Accept-Language': 'en-US,en;q=0.9',
    'Accept-Encoding': 'gzip, deflate, br',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
    'Referer': 'https://www.google.com/',
    'Connection': 'keep-alive'
}

res = requests.get(url, headers)
res
```

```python
res.text
```

```python
# create beautifulSoup object
soup = BeautifulSoup(res.text, 'lxml')
```

```python
# format html file
print(soup.prettify())
```

```python
# Finds all h1 tags
soup.find_all('h1')
```

```python
# extract text from h1 tag
soup.find_all('h1')[0].text
```

```python
# Find all the h2 tags
soup.find_all('h2')
```

```python
# Extract names and save in a list
title = []
for i in range(1, 21):
    heading = soup.find_all('h2')[i].text
    title.append(heading)
    
print(title)
```

```python
# To extract content from specified class
soup.find_all('h3', class_='class_name')
```

```python
# find all p tags
len(soup.find_all('p'))
```

```python
# Extract description and save in a list
review = []
for i in range(1, 23):
    if(i!=2 and i!=3):
        rev = soup.find_all('p')[i].text
        review.append(rev)

print(review)
```

```python
# create dataframe from the lists
d = {'Name':title, 'Description':review}
df = pd.DataFrame(d)
df.head(5)
```

```python
df.to_csv('best_companies.csv')
```
