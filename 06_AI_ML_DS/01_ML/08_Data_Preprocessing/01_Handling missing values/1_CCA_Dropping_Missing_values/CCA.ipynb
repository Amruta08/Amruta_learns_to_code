# Handling Missing values
  1. Remove them (CCA - Complete case analysis)
  2. Impute them :-
     * Univariate :- Can be performed using sklearn SimpleImputer() class
        * Numerical
           * Mean/Median
           * Random value
           * End of distribution
           * Arbitary
         * Categorical
           * mode
           * Write NaN
      * Multivariate
          * KNN Imputer
          * Iterative Imputer (MICE)
  3. Missing Indicator

#CCA

```python
import pandas as pd
import numpy as np
df = pd.read_csv('data_science_job.csv')
df.head()
```

```python
# Find % of missing values for each column
df.isnull().mean()*100
```

```python
# Filtering columns which have less than 5% missing values
cols = [var for var in df.columns if df[var].isnull().mean()<0.05 and df[var].isnull().mean()>0]
cols
```

```python
# Finding percentage of data lost after performing CCA
len(df[cols].dropna())/len(df) #11% data is lost
```

```python
# Performing CCA
new_df = df[cols].dropna()
df.shape, new_df.shape
```

```python
# Before and After hist plot of training_hours overlayed on top of each other
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
df['training_hours'].hist(bins=50, ax=ax, density=True, color='red')
new_df['training_hours'].hist(bins=50, ax=ax, density=True, color='blue')
```

```python
# Before and after kde plots for training_hours
fig = plt.figure()
ax = fig.add_subplot(111)
df['training_hours'].plot.density(color='red')
new_df['training_hours'].plot.density(color='green')
```
