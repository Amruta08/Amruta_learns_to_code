```python
# Load dataset
import numpy as np
import pandas as pd

df = pd.read_csv('weight-height.csv')
df.head()
```

```python
# Plotting kdeplot of height
import seaborn as sns
sns.kdeplot(df['Height'])
```

```python
# Plotting boxplot of height
sns.boxplot(x=df['Height'])
```

```python
# Viewing statisitcs of height
df['Height'].describe()
```

## Trimming - Removing outliers

```python
# Trimming
lower, upper = df['Height'].quantile([0.01, 0.99])

new_df = df[(df['Height']<=upper) & (df['Height']>=lower)]
```

```python
new_df['Height'].describe()
```

```python
sns.kdeplot(new_df['Height'])
```

```python
sns.boxplot(x=new_df['Height'])
```

## Capping outliers :- Replacing outliers

```python
# Capping
lower, upper = df['Height'].quantile([0.01, 0.99])

df['Height'] = df['Height'].clip(lower, upper)
```

```python
df['Height'].describe()
```

```python
sns.kdeplot(df['Height'])
```

```python
sns.boxplot(x=df['Height'])
```
