```python
# Data is divided into 2 types :-
# 1. Categorical data
#    Are of further 2 types :-
#        a. nominal (no rel. or ordering among data), Eg.  gender -> one hot encoding
#        b. ordinal (there exits ordering among data) Eg. rank, review -> ordinal encoding
#        c. label encoding :- similar like ordinal encoding, but designed for the output column
# 2. Numerical data
```

##Loading dataset

```python
import pandas as pd
import numpy as np

df = pd.read_csv('customer.csv')
df.sample(5)
```

```python
# We dont need the first 2 columns
df = df.iloc[:,2:]
df.head(3)
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
X = df.iloc[:,0:2]
Y = df.iloc[:,-1]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

## Applying Ordinal Encoding on review and education

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
X_train = oe.fit_transform(X_train)
X_test = oe.transform(X_test)
```

```python
oe.categories_
```

## Applying Label Encoding on Purchase column

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit_transform(Y_train)
le.transform(Y_test)
```
