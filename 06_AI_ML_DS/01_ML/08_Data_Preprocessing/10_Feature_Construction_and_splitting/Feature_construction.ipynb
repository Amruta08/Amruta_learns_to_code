```python
# Load dataset
import numpy as np
import pandas as pd

df = pd.read_csv('train.csv', usecols=['Age','Pclass','SibSp','Parch','Survived'])
df.head()
```

```python
# Drop missing values
df.dropna(inplace=True)
```

```python
# Applying feature construction (SibSp = Siibling spouse, Parch = Parent child)
df['Family_size'] = df['SibSp'] + df['Parch'] + 1
df.head()
```

```python
# Custom function to define family type
def FamilyType(num):
  if num == 1:
    #alone
    return 0
  elif num > 1 and num <= 4:
    # small family
    return 1
  else:
    return 2

df['Family_type'] = df['Family_size'].apply(FamilyType)
df.sample(5)
```

```python
# Dropping SibSp, Parch, Family_size
df.drop(['SibSp', 'Parch', 'Family_size'], axis=1, inplace=True)
df.head()
```
