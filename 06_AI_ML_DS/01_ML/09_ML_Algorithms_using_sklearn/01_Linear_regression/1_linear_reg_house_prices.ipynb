```python
# Load Dataset
import pandas as pd
import numpy as np
df = pd.read_csv('dataset.csv')
df
```

```python
# Plot Scatterplot
import matplotlib.pyplot as plt
plt.scatter(df.area, df.price)
plt.xlabel('area (sq ft)')
plt.ylabel('price (US $)')
```

```python
# Applying linear regression
from sklearn.linear_model import LinearRegression
X = df[['area']]
Y = df['price']
reg = LinearRegression()
reg.fit(X,Y)
```

```python
# Predict price of home whose area is 3300 square ft.
reg.predict([[3300]])
```

```python
# y=mx+c
# price = m*area + c

# Value of m
print("m :- ", reg.coef_)

# value of c
print("c :- ", reg.intercept_)
```

```python
# plot regression line
plt.scatter(df.area, df.price)
plt.plot(df.area, reg.predict(df[['area']]), color='red')
plt.xlabel('area (sq ft)')
plt.ylabel('price (US $)')
```

## Applying our reg model to predict areas house prices from area.csv

```python
# Load areas.csv
areas = pd.read_csv('areas.csv')
areas.head()
```

```python
# Predict prices of areas present in areas.csv dataset
predicted_prices = reg.predict(areas)
print(predicted_prices)
```
