```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
```

```python
df = pd.read_csv("test_scores.csv")
df
```

```python
math_score = np.array(df.math)
math_score
```

```python
cs_score = np.array(df.cs)
cs_score
```

```python
def gradient_descent(x, y):

    # Creating a scatter plot for x & y co-ordinates
    plt.scatter(x, y, linewidth=5)

    # Initial state of slope and intercept
    m_curr = b_curr = 0;

    iterations = 10000
    n = len(x)
    learning_rate = 0.0001

    cost_previous = 0;

    for i in range(iterations):

        # Calculating y_predicted
        y_predicted = m_curr*x + b_curr

        # Plot the y_predicted line
        plt.plot(x, y_predicted, color="green")

        # Calculating cost as Mean Sqaured error
        cost = (1/n)*sum([val**2 for val in (y-y_predicted)])

        # Calculating d/dm & d/db
        md = -(2/n)*sum(x*(y-y_predicted))
        bd = -(2/n)*sum(y-y_predicted)

        # Calculating new slope and intercept
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd

        if(math.isclose(cost,cost_previous,rel_tol=1e-20)):
            break;
        cost_previous = cost

        print(f"m {m_curr}, cost {cost}, b {b_curr}, iteration {i}")

    return m_curr, b_curr
```

```python
m, b = gradient_descent(math_score, cs_score)
```
