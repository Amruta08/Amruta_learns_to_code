```python
# Load dataset
import pandas as pd
df = pd.read_csv('salaries.csv')
df.head()
```

```python
# Splitting data into input and output
X = df.drop('salary_more_then_100k', axis=1)
Y = df['salary_more_then_100k']
```

```python
# Applying ordinal encoding
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder()

X['company'] = oe.fit_transform(X[['company']])
X['job'] = oe.fit_transform(X[['job']])
X['degree'] = oe.fit_transform(X[['degree']])
X.head()
```

```python
# Applying decison tree algorithm
from sklearn.tree import DecisionTreeClassifier, plot_tree
model = DecisionTreeClassifier()
model.fit(X,Y)
```

```python
# Predict Is salary of Google, Computer Engineer, Bachelors degree > 100 k ?
model.predict([[2, 1, 0]])
```

```python
# Predict Is salary of Google, Computer Engineer, Masters degree > 100 k ?
model.predict([[2, 1, 1]])
```

```python
# Plot decsion tree
plot_tree(model)
```
