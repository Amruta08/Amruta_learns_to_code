```python
# Load dataset
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
# Adding target column and target_name column to dataset
df['target'] = iris.target
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
# Spliting data based on flower name
df0 = df[df.target == 0]
df1 = df[df.target == 1]
df2 = df[df.target == 2]
```

```python
# Visualizing data on based Sepal length vs Sepal Width
import matplotlib.pyplot as plt

plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color='green')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color='red')
plt.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'], color='blue')
```

```python
# Visualizing data on based Petal length vs Petal Width
import matplotlib.pyplot as plt

plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'], color='green')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color='red')
plt.scatter(df2['petal length (cm)'], df2['petal width (cm)'], color='blue')
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X = df.drop(['target', 'flower_name'], axis=1)
Y = df['target']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
# Apply SVM
from sklearn.svm import SVC
model = SVC(C=10, gamma='auto', kernel='linear') # C stands for regularization
model.fit(X_train, Y_train)
```

```python
# Check accuracy
model.score(X_test, Y_test)
```
