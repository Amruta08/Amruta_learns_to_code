```python
# Load dataset
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
df.head()
```

```python
# Displaying scatterplot between features
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 5))

plt.subplot(1,2,1)
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='target', data=df)

plt.subplot(1,2,2)
sns.scatterplot(x='petal length (cm)', y='petal width (cm)', hue='target', data=df)
```

```python
# Train test split
from sklearn.model_selection import train_test_split
X = df.drop('target', axis=1)
Y = df['target']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
# Applying KNN algo and checking accuracy
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, Y_train)
knn.score(X_test, Y_test)
```

```python
# Plot cofusion matrix
from sklearn.metrics import confusion_matrix
Y_pred = knn.predict(X_test)
cm = confusion_matrix(Y_test, Y_pred)
sns.heatmap(cm, annot=True)
```

```python
# Display classification report
from sklearn.metrics import classification_report
print(classification_report(Y_test, Y_pred))
```
