```python
# Load dataset
import numpy as np
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
df = pd.DataFrame(digits.data, columns=digits.feature_names)
df['target'] = digits.target
df.head()
```

```python
# Train test split
from sklearn.model_selection import train_test_split

X = df.drop('target', axis=1)
Y = df['target']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
# Scaling X_train and Y_test
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
# Applying logistic regression and checking accuracy
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=200)
lr.fit(X_train, Y_train)
lr.score(X_test, Y_test)
```

## PCA decompostion part

### Maintaining 95% variance

```python
# PCA decompostion such that 95% of variance is retained
from sklearn.decomposition import PCA
pca = PCA(0.95)
X_pca = pca.fit_transform(X)
```

```python
# Shape before and after
print("Shape before PCA :- ", X.shape)
print("Shape after PCA :- ", X_pca.shape)
```

```python
# Variance ratio for each component
pca.explained_variance_ratio_
```

```python
# No. of components choosen
pca.n_components_
```

```python
# Now checking accuracy of our model
X_train_pca, X_test_pca, Y_train_pca, Y_test_pca = train_test_split(X_pca, Y, test_size=0.2)
lr.fit(X_train_pca, Y_train_pca)
lr.score(X_test_pca, Y_test_pca)
```

###Specifying no. of components

```python
# PCA decompostion by specifying no. of components
pca = PCA(n_components=10)
X_pca = pca.fit_transform(X)
```

```python
# Shape before and after
print("Shape before PCA :- ", X.shape)
print("Shape after PCA :- ", X_pca.shape)
```

```python
# Variance ratio for each component
pca.explained_variance_ratio_
```

```python
# Now checking accuracy of our model
X_train_pca, X_test_pca, Y_train_pca, Y_test_pca = train_test_split(X_pca, Y, test_size=0.2)
lr.fit(X_train_pca, Y_train_pca)
lr.score(X_test_pca, Y_test_pca)
```
