# Remove HTML Tags

#### Regular expression (r'<.*?>') explanation :-
* `<` and `>` → Starts with `<` and ends with `>`
* `.` → Matches any character
* `*` → Matches 0 or more characters
* `?` → Makes `*` lazy instead of greedy and makes `*` stop after the first closing `>` instead of going too far


```python
# Remove's all html tags from text
import re
def remove_html(data):
    p = re.compile(r'<.*?>')
    return p.sub('', data)

sample_text = '<h1>HTML element</h1> <p>para element</p>'
remove_html(sample_text)
```

```python
# Remove html tags from review column
df['review'] = df['review'].apply(remove_html)
df
```

# Remove URL's

#### Regular expression (r'https?://\S+|www\.\S+') explanation :-
* `https?://` → Matches `http://` or `https://` (`s?` means `s` is optional)
* `\S+` → Matches one or more non-space characters (the rest of the URL)
* `|` → OR operator, means match either the left or right pattern
* `www\.\S+` → Matches URLs starting with `www.` followed by non-space characters


```python
def remove_url(text):
    p = re.compile(r'https?://\S+|www\.\S+')
    return p.sub(r'',text)

text1 = 'Google search here www.google.com'
remove_url(text1)
```

# Remove Punctuation

```python
import string
string.punctuation
```

```python
# Slow approach
import string
exclude = string.punctuation

def remove_punc(text):
    for char in exclude:
        text = text.replace(char, '')
    return text

text2 = 'string. With. Punctuation?'
remove_punc(text2)
```

```python
# Fast and better approach
def remove_punc_fast(text):
    return text.translate(str.maketrans('', '', exclude))

text3 = 'string. With. Punctuation?'
remove_punc(text3)
```

```python
# removing punctuation from tweet column
df1['tweet'] = df1['tweet'].apply(remove_punc_fast)
df1
```

```python
from nltk.corpus import stopwords

def remove_stopwords(text):
    new_text = []

    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    return " ".join(new_text)

remove_stopwords('probably my all-time favorite movie')
```

# Handling emojis

```python
# Unicode normalization
emoji_text = "Today was a good day 😊, Today was a sunny day☀️"
emoji_text.encode('utf-8')
```

```python
# Remove emojis
import re
def remove_emoji(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"  # Dingbats (stars, scissors, hearts)
                           u"\U000024C2-\U0001F251"  # Enclosed characters, other symbols
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

remove_emoji("I liked the movie 😘")
```

# Stemming and lemmatization
- In Grammar, inflection is the modification of a word to express different grammatical categories such as tense, case, voice, aspect, person, number, gender and mood.
- Stemming is the process of reducing inflection in words to their root forms such as mapping a group of words to the same stem even if the stem itself is not a valid word in the language.
- Lemmatization, unlike stemming, reduces the inflected words properly ensuring that the root word belongs to the language. In Lemmatization root word is called Lemma. A lemma is the canonical form, dictionary form, or citation form of a set of words.

## Stemming

```python
# Converting words into thier root form E.g. Walking into Walk
# However some root words may not be valid words in the language

# Stemming using nltk PorterStemmer
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
def stem_words(text):
    return " ".join([ps.stem(word) for word in text.split()])

sample = "walk walks walking walked movie story"
stem_words(sample)
```

## Lemmatization

```python
import nltk
nltk.download('wordnet')
```

```python
# Converts root words into thier root form ensuring the root word belongs to the language
# root word is called lema here
# Slower than stemming

# Lemmatization using wordnet lemmatizer
# wordnet is a lexical dictionary
import nltk
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

sentence = "He was running and eating at same time. He has bad habit of swimming after playing long hours in the Sun."
punctuations="?:!.,;"
sentence_words = nltk.word_tokenize(sentence)

for word in sentence_words:
    if word in punctuations:
        sentence_words.remove(word)

for word in sentence_words:
    print(wordnet_lemmatizer.lemmatize(word, pos='v'))   #pos='v' -> implies parts of speech as verb
```
