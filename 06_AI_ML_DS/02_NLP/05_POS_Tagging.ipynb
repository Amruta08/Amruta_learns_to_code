```python
import spacy
nlp = spacy.load('en_core_web_sm')  # Load english language dictionary
```

```python
doc = nlp("I will google about facebook")
```

```python
doc.text
```

```python
doc[2]
```

```python
# Coarse POS
doc[2].pos_
```

```python
# Fine grained POS
print(doc[2].tag_)
print(spacy.explain('VB'))
```

```python
# Display POS of the given sentence
for word in doc:
  print(word.text, "-->", word.pos_, "|", word.tag_, "|",spacy.explain(word.tag_))
```

```python
doc2 = nlp(u"I left the room")
for word in doc2:
  print(word.text, "-->", word.pos_, "|", word.tag_, "|",spacy.explain(word.tag_))
```

```python
doc6 = nlp(u"The quick brown fox jumped over the lazy dog")
for word in doc6:
  print(word.text, "-->", word.pos_, "|", word.tag_, "|",spacy.explain(word.tag_))
```

```python
# Visulize the POS
from spacy import displacy
displacy.render(doc6, style="dep")
```
